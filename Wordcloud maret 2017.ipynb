```python
from os import path
from wordcloud import WordCloud,STOPWORDS
```

```python
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
```

```python
text = open(path.join(d, 'D:/DIGITALENT/Aggie/3. Uploaded Data/wordcloud maret 2017.txt')).read()
stopwords = set(STOPWORDS)
stopwords.add("di")
stopwords.add("non")
stopwords.add("ini")
stopwords.add("ke")
stopwords.add("tidak")
stopwords.add("dari")
stopwords.add("sudah")
stopwords.add("sejak")
stopwords.add("lama")
stopwords.add("ada")
stopwords.add("sebelum")
stopwords.add("untuk")
stopwords.add("telah")
stopwords.add("dan")
stopwords.add("dia")
stopwords.add("tentang")
stopwords.add("hal")
stopwords.add("yang")
stopwords.add("dengan")
stopwords.add("asal")
stopwords.add("sebut")
stopwords.add("kepada")
stopwords.add("karena")
stopwords.add("diduga")
stopwords.add("diyakini")
stopwords.add("masih")
stopwords.add("island")
stopwords.add("tahun")
stopwords.add("belasan")
stopwords.add("sekitar")
stopwords.add("berikan")
stopwords.add("minta")
stopwords.add("okezone")
stopwords.add("harus")
```

```python
text = open(path.join(d, 'D:/DIGITALENT/Aggie/3. Uploaded Data/wordcloud maret 2017.txt')).read()
wordcloud = WordCloud(width=5000, height=2700,background_color='white',colormap='Dark2',stopwords=stopwords).generate(text)
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
